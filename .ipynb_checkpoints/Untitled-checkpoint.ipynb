```python
#Import packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
from scipy import stats
```

```python
%load_ext autoreload
%autoreload 2

deg="\u2103"
#Tilda for n code used in plot titles
accent='\u0303'
n='n'
nt=n+accent #N Tilda
```

```python
group_number = 2
path_dir = Path.cwd() / 'data' / f'{group_number}'
# examine available files in your folder
list(path_dir.glob('*gz'))
```

```python
# specify the site name
name_of_site = 'US-MMs'
```

```python
# load dataset
site_file = name_of_site + '_clean.csv.gz'
path_data = path_dir / site_file
df_data = pd.read_csv(path_data, index_col='time', parse_dates=['time'])
df_data.head()
```

```python
df_fluxes = df_data.filter(['NETRAD', 'H', 'LE'])
```

```python
#Allocating each year DJF
df_1999DJF=df_fluxes.loc['1999 01 01':'1999 02 28'] 
df_2000DJF=df_fluxes.loc['1999 12 01':'2000 02 29']
df_2001DJF=df_fluxes.loc['2000 12 01':'2001 02 28']
df_2002DJF=df_fluxes.loc['2001 12 01':'2002 02 28']
df_2003DJF=df_fluxes.loc['2002 12 01':'2003 02 28']
df_2004DJF=df_fluxes.loc['2003 12 01':'2004 02 29']
df_2005DJF=df_fluxes.loc['2004 12 01':'2005 02 28']
df_2006DJF=df_fluxes.loc['2005 12 01':'2006 02 28']
df_2007DJF=df_fluxes.loc['2006 12 01':'2007 02 28']
df_2008DJF=df_fluxes.loc['2007 12 01':'2008 02 29']
df_2009DJF=df_fluxes.loc['2008 12 01':'2009 02 28']
df_2010DJF=df_fluxes.loc['2009 12 01':'2010 02 28']
df_2011DJF=df_fluxes.loc['2010 12 01':'2011 02 28']
df_2012DJF=df_fluxes.loc['2011 12 01':'2012 02 29']
df_2013DJF=df_fluxes.loc['2012 12 01':'2013 02 28']
df_2014DJF=df_fluxes.loc['2013 12 01':'2014 02 28']
df_2015DJF=df_fluxes.loc['2014 12 01':'2015 02 28']
df_2016DJF=df_fluxes.loc['2015 12 01':'2016 02 29']
df_2017DJF=df_fluxes.loc['2016 12 01':'2017 02 28']
df_2018DJF=df_fluxes.loc['2017 12 01':'2018 02 28']
df_2019DJF=df_fluxes.loc['2018 12 01':'2019 02 28']
```

```python
#ONI 
data=np.genfromtxt("Nino A2.csv", delimiter=',', filling_values=np.nan)
#year=np.genfromtxt("Nino A2.csv", delimiter=',', usecols=(0), dtype=None)
year=data[:,0]
SST=data[:,1]
#assign variables
x=year
y=SST
x[0]=1999
```

```python
#for GRAPH
y1=[]
for i in range(0,21):
    y1.append(0)
    
#plotting
fig=plt.figure(figsize=(20,10))
ax=fig.add_subplot(1,1,1)
ax.spines['right'].set_visible(True)
ax.spines['top'].set_visible(True)
ax.plot(x, SST, color='gray')
#Colouring above and below 0
ax.fill_between(x, SST, y1, where=SST >= y1, facecolor='red', interpolate=True, alpha=0.8)
ax.fill_between(x, SST, y1, where=SST <= y1, facecolor='blue', interpolate=True, alpha=0.8)

#labels
ax.set_title('Time Series of Oceanic Ni'+str(nt)+'o Index 3.4', fontsize=20)
ax.set_xlabel('Time (Years)', fontsize=18)
ax.set_ylabel('Oceanic Ni'+str(nt)+'o Index 3.4 ('+str(deg)+')', fontsize=18)

#axis numbers
major_ticks = np.arange(2000,2021,5)
minor_ticks = np.arange(1999,2021,1)

ax.set_xticks(major_ticks)
ax.set_xticks(minor_ticks, minor=True)
ax.set_xticklabels(major_ticks,fontsize=16)
ax.set_yticks(np.arange(-2.5, 3, 0.25), minor=True) 
ax.set_yticks(np.arange(-2.5, 3, 0.5))
ax.set_yticklabels(np.arange(-2.5, 3, 0.5), fontsize=16)

#Adding threshold lines
l1=plt.axhline(0.5,color='black',ls='--', linewidth=0.7)
l2=plt.axhline(-0.5,color='black',ls='--',linewidth=0.7)
l3=plt.axhline(0,color='black',ls='-', linewidth=0.7 ) #0 axis
l1.set_label('+/-0.5'+str(deg)+' Threshold')
plt.legend(loc='lower right', fontsize=17)
#grid
# And a corresponding grid
ax.grid(which='both')

# Or if you want different settings for the grids:
ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)

plt.show()



#Finding years
#using 0.5 as threshold
#EN years

EN_years=[]
for i in range(0,21):
    if SST[i]>0.5:
        EN_years.append(year[i])
    else:
        i=i+1
        
#La Nina years
LN_years=[]
for i in range(0,21):
    if SST[i]<-0.5:
        LN_years.append(year[i])
    else:
        i=i+1

print(EN_years)
print(LN_years)
```

```python
#MEAN OF RATIOS EACH YEAR
YEARLY_QH_RATIO=[]
NR_99=df_1999DJF[df_1999DJF.NETRAD!=0]
np.mean(df_1999DJF.H/NR_99.NETRAD)
YEARLY_QH_RATIO.append(np.mean(df_1999DJF.H/NR_99.NETRAD))
NR_00=df_2000DJF[df_2000DJF.NETRAD!=0]
np.mean(df_2000DJF.H/NR_00.NETRAD)
YEARLY_QH_RATIO.append(np.mean(df_2000DJF.H/NR_00.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2001DJF.H/df_2001DJF.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2002DJF.H/df_2002DJF.NETRAD))
NR_03=df_2003DJF[df_2003DJF.NETRAD!=0]
np.mean(df_2003DJF.H/NR_03.NETRAD)
YEARLY_QH_RATIO.append(np.mean(df_2003DJF.H/NR_03.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2004DJF.H/df_2004DJF.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2005DJF.H/df_2005DJF.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2006DJF.H/df_2006DJF.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2007DJF.H/df_2007DJF.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2008DJF.H/df_2008DJF.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2009DJF.H/df_2009DJF.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2010DJF.H/df_2010DJF.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2011DJF.H/df_2011DJF.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2012DJF.H/df_2012DJF.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2013DJF.H/df_2013DJF.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2014DJF.H/df_2014DJF.NETRAD))
YEARLY_QH_RATIO.append(np.mean(df_2015DJF.H/df_2015DJF.NETRAD))
#YEARLY_QH_RATIO.append(np.mean(df_2016DJF.H/df_2016DJF.NETRAD))
#YEARLY_QH_RATIO.append(np.mean(df_2017DJF.H/df_2017DJF.NETRAD))
#YEARLY_QH_RATIO.append(np.mean(df_2018DJF.H/df_2018DJF.NETRAD))
#YEARLY_QH_RATIO.append(np.mean(df_2019DJF.H/df_2019DJF.NETRAD))
year=year[0:17]
SST=SST[0:17]
plt.scatter(SST,YEARLY_QH_RATIO)
```

```python
fig=plt.figure(figsize=(14,7))
ax = fig.add_subplot(111)
#points
for i in range(17):
    x = SST[i]
    y = YEARLY_QH_RATIO[i]
    plt.scatter(x, y, marker='o', color='red')
    plt.text(x+0.01, y+0.01, int(year[i]), fontsize=10,verticalalignment='baseline', horizontalalignment='center')
#line of best fir
slope, intercept, r_value, p_value, std_err = stats.linregress(SST,YEARLY_QH_RATIO)    
line = slope*SST+intercept
ax.plot(SST, line, color='blue', label='Regression Line')
plt.legend(loc='lower right',fontsize=11)

plt.title('Correlation between $Q_H$/$Q*$ and Oceanic Ni'+str(nt)+'o Index 3.4', fontsize=17)    
plt.xlabel('Oceanic Ni'+str(nt)+'o Index 3.4 ('+str(deg)+')', fontsize=14)
ax.set_ylabel('$Q_H/Q*$', fontsize=14)

ax.grid()
plt.axvline(x=0.5,linestyle='--', color='k', linewidth=1)
plt.axvline(x=-0.5,linestyle='--', color='k', linewidth=1, label='Ni'+str(nt)+'o Threshold')
plt.legend()
print('QH RVALUE', r_value)
print(r_value**2)
print(p_value)
```
